- PassengerID: 각 승객에게 주어진 고유 ID 번호  
- Survived: 생존 여부(0=사망, 1=생존) => Target 데이터(종속변수)->y=ax+bz+c ->독립변수,데이터

- SibSp: 형제자매/배우자의 수  
- Parch: 함께 탑승한 부모님/아이들의 수  
- Ticket: 티켓 번호  
- Fare: 탑승 요금  
- Cabin: 수하물 번호 또는 객실 카테고리   
- Embarked: 선착장(C=Cherbourg, Q=Queenstown,S=Southampton)

```python
result = df.add(1)
result
```

```python
result = df+1
result
```

```python
data2 = [[3],[4],[5]]
df2 = pd.DataFrame(data=data2,index=['row1','row2','row3'],columns=['col1'])
df2
```

```python
result = df.add(df2)
result
```

```python
result = df.add(df2,fill_value=0)     # df2에 없는 컬럼에 0을 채워서 연산
result
```

```python
result = df.add(1)  # sub(), mul(), div(), 나머지연산 mod()
result
```
