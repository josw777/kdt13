# Merge

```python
df1 = pd.DataFrame({'고객번호':[1001,1002,1003,1004,1005,1006,1007],
                    '이름': ['둘리','도우너','또치','길동','희동','마이콜','영희']},
                   columns=['고객번호','이름'])
df1
```

```python
df2 = pd.DataFrame({'고객번호':[1001,1001,1005,1006,1008,1001],
                    '금액': [10000,20000,15000,5000,10000,30000]},
                   columns=['고객번호','금액'])
df2
```

```python
df3 = pd.merge(df1,df2)    # 기본: inner, 교집합과 비슷
df3
```

```python
df4 = pd.merge(df1,df2,how='outer')
df4
```

```python
df5 = pd.merge(df1,df2,how='left')
df5
```

```python
df6 = pd.merge(df1,df2,how='right')
df6
```

```python
df1 = pd.DataFrame({
    '품종':['setosa','setosa','virginica','virginica'],
    '꽃잎길이':[1.4,1.3,1.5,1.3]},
    columns=['품종','꽃잎길이'])
df1
```

```python
df3 = pd.merge(df1,df2)
df3
```

```python
df1 = pd.DataFrame({
    '고객명':['춘향','춘향','몽룡'],
    '날짜':['2018-01-01','2018-01-02','2018-01-01'],
    '데이터':['20000','30000','100000']})
df1
```

```python
df2 = pd.DataFrame({
    '고객명':['춘향','몽룡'],
    '데이터':['여자','남자']})
df2
```

```python
df3 = pd.merge(df1,df2,on='고객명')    # on : 기준열을 지정, 서로다른 성격의 데이터열이 중복될때
df3
```

```python
df1 = pd.DataFrame({
    '이름':['춘향','춘향','몽룡'],
    '날짜':['2018-01-01','2018-01-02','2018-01-01'],
    '데이터':['20000','30000','100000']})
df1
```

```python
df2 = pd.DataFrame({
    '성명':['춘향','몽룡'],
    '데이터':['여자','남자']})
df2
```

```python
df3 = pd.merge(df1,df2,left_on='이름',right_on='성명')
df3
```
