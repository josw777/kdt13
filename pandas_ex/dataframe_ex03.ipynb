### pivot_table
형식) pivot_table(data, values=None, index=None, columns=None, aggfunc='mean', fill_value=None, margins=False, margins_name='All')
- data: 분석할 데이터프레임 (메서드일 때는 필요하지 않음)
- values: 분석할 데이터프레임에서 분석할 열
- index: 행 인덱스로 들어갈 키 열 또는 키 열의 리스트
- columns: 열 인덱스로 들어갈 키 열 또는 키 열의 리스트
- aggfunc: 분석 메서드 지정
- fill_value: NaN 대체 값 설정
- margins: 오른쪽과 아래에 합계를 붙일지 여부 지정
- margins_name: 합계 열(행)의 이름 지정

```python
df2 = pd.DataFrame({'key1':['A','A','B','B','A'],
                    'key2':['one','two','one','two','one'],
                    'data1':[1,2,3,4,5],
                    'data2':[10,20,30,40,50]})
df2
```

```python
df2.pivot_table('data1','key1','key2')  #default aggfunc=mean
```

```python
df2.pivot_table('data1','key1','key2',aggfunc=sum)
```

```python
df2.pivot_table('data1','key1','key2',margins=True)
```

```python
df2.pivot_table('data1','key1','key2',margins=True,margins_name='평균')
```

```python
df2.pivot_table('data1',index=['key1','key2'])
```

```python
np.random.seed(0)
df1 = pd.DataFrame(np.random.randn(100,3),
                   index=pd.date_range('1/1/2018',periods=100),
                   columns=['A','B','C']).cumsum()
df1.tail(3)
```

```python
df1.plot(kind='line')    # default는 line차트   라인차트는 시계열데이터에 사용 시간에 따른 변화를 한눈에 파악
plt.title('Pandas의 Plot메소드 사용 예')
plt.xlabel('시간')
plt.ylabel('Data')
plt.show()
# 축: tick, 축값: ticklabel
```

```python
iris = sns.load_dataset('iris')  # 아이리스(붓꽃) 데이터,   titanic: 타이타닉데이터,    tips: 팁데이터
iris
```

꽃받침 길이, 꽃받침 넓이, 꽃잎 길이, 꽃잎 넓이로 품종을 분류하는 데이터

```python
iris.sepal_length[:20].plot(kind='bar',rot=0)
plt.title('꽃받침의 길이 시각화')
plt.xlabel('Data')
plt.ylabel('꽃받침의 길이')
plt.show()
```

```python
iris[:5].plot.bar(rot=0)
plt.title('Iris 데이터의 Bar Plot')
plt.xlabel('Data')
plt.ylabel('각 Feature의 값')
plt.ylim(0, 7)
plt.show()
```

```python
df2 = iris.groupby(iris.species).mean()
df2.columns.name = 'feature'
df2
```

```python
df2.plot.bar(rot=0)
plt.title('각 종의 Feature별 평균')
plt.xlabel('평균')
plt.ylabel('종')
plt.ylim(0, 8)
plt.show()
```
