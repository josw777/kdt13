```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.family']='Malgun Gothic'
plt.rcParams['axes.unicode_minus']=False

iris = sns.load_dataset('iris')
iris
```

(1) iris 데이터에서 species별 sepal_length의 평균을 구한 후 이를 'bar' 타입으로 그려보세요.

```python
df = iris.groupby('species')['sepal_length'].min()
df
```

```python
df.plot(kind='bar',rot=0)
plt.title('species별 sepal_length의 평균')
plt.xlabel('species')
plt.ylabel('sepal_length 평균')
plt.show()
```

(2) iris 데이터에서 품좀별 데이터의 개수를 파이 차트 형태로 그려보세요.

```python
df = iris.groupby('species').size()
df
```

```python
df.plot.pie(autopct='%2d%%')
plt.title('품종별 데이터의 개수')
plt.axis('equal')
plt.show()
```
