[문제2] 아래의 코드를 사용하여 sample2 중 160 이하인 데이터만 필터링여 출력하는  코드를 작성하시오.

```python
import pandas as pd
import numpy as np

np.random.seed(20)
sample2 = pd.Series(np.random.randint(100,200,size=(15,)))
sample2
```

```python
a = sample2[sample2 <= 160]
a
```

[문제3] 다음과 같은 Series를 생성하는 코드를 작성하시오.

```python
sample3 = Series(['apple',np.nan,'banana','kiwi','gubong'],['가','나','다','라','마'])
sample3
```

[문제4] 위 예제에서 결측치가 아닌 데이터만 필터링하여 출력하는 코드를 작성하시오.

```python
sample4 = sample3[sample3.notnull()]
sample4
```

[문제5] 위 예제에서 결측치 데이터만 필터링하여 출력하는 코드를 작성하시오.

```python
sample5 = sample3[sample3.isnull()]
sample5
```
