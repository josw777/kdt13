# Series 클래스

```python
import pandas as pd
```

```python
s = pd.Series([1,3,5,7,9])
print(s)
```

```python
print(s.index)
```

```python
print(s.values)
```

```python
s = pd.Series([9904312, 3448737, 2890451, 2466052],
              index=["서울","부산","인천","대구"])
print(s)
```

```python
print(s.index)
```

```python
print(s.values)
```

```python
s.name = "인구"
s.index.name = "도시"
print(s)
```

```python
s.index.name = None
print(s)
```

```python
print(pd.Series(range(1,5)))
```

```python
s= pd.Series(['samsung',8100])
print(s)
```

타입이 섞여있으면 object타입으로 됨
계산하려면 형변환필요

```python
from pandas import Series

data = [1000,2000,3000]
s = Series(data)        
print(s.index)
print(s.index.to_list())
```

```python
data = [1000, 2000, 3000]
s = Series(data)
s.index = ['메로나', '구구콘', '하겐다즈']
print(s)
```

```python
data = [1000, 2000, 3000]
index = ['메로나', '구구콘', '하겐다즈']
s = Series(data = data, index=index)          # s = Series(data,index)    # s = Series(index=index, data=data)
print(s)
```

```python
s2 = s.reindex(['메로나','비비빅','구구콘'])
print(s2)
```

중간에 nan이 들어가면 float로 바뀜

```python
s2 = s2.fillna(0)
print(s2)
```

```python
price = [42500, 42550, 41800, 42550, 42650]
date = ['2019-05-31', '2019-05-30', '2019-05-29', '2019-05-28', '2019-05-27']
s = Series(price,date)
print(s)
# 시계열 데이터
```

```python
data = {
    '2019-05-31' : 42500,
    '2019-05-30' : 42550,
    '2019-05-29' : 41800,
    '2019-05-28' : 42550,
    '2019-05-27' : 42650
}
s = Series(data)
print(s)
```

```python
print(s.index)
```

```python
print(s.index.dtype)
```

```python
print(s.values)
```

# 인덱싱

```python
s1 = Series([10, 20, 30])
s2 = Series([10, 20, 30], index = [1, 2, 3])
print(s1[0])
print(s2[1])
```

```python
s3 = Series([10, 20, 30], index = ['a', 'b', 'c'])
print(s3['a'])
```
