# 예제

```python
def remove_comma(x):
    return int(x.replace(',',''))
```

```python
s = Series(['1,234','5,678','9,876'])
result = list(s.map(remove_comma))
print(result)
```

# 정렬

```python
from pandas import Series
data = [3.1,2.0,10.1,5.1]
index = ['000010','000020','000030','000040']
s = Series(data=data, index=index)
print(s)
```

```python
s1 = s.sort_values(ascending=False)
print(s1)
```

```python
s2 = s.rank(ascending=False)    #ascending=True가 기본값
print(s2)
```
